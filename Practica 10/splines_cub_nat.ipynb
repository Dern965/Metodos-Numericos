```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
```

```python
"""
    Calcula los coeficientes del spline cúbico natural.

    Parámetros:
    x: Lista con las coordenadas x de los puntos de datos.
    y: Lista con las coordenadas y de los puntos de datos.

    Retorna:
    Objeto CubicSpline que representa el spline cúbico natural.
"""
def SplineCubico_natural(x, y):
    #El parametro  bc_type='natural' hace referencia a que se esta utlizando la interpolacion cúbica natural.
    CubicSN = CubicSpline(x, y, bc_type='natural')
    return CubicSN


"""
    Funcion para graficar el spline cúbico natural junto con los puntos de los datos.

    Parámetros:
    OCubicSpline: Objeto CubicSpline.
    x: Lista con las coordenadas x de los puntos de datos.
    y: Lista con las coordenadas y de los puntos de datos.
"""
def spline_cubNat_graf(OCubicSpline, x, y):
    plt.figure(figsize=(8, 6))
    plt.scatter(x, y, color='yellow', label='Datos')
    x_scn = np.linspace(x[0], x[-1], 100)
    y_scn = OCubicSpline(x_scn)
    plt.plot(x_scn, y_scn, label='Spline Cúbico Natural')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Spline Cúbico Natural')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python

# Pedir al usuario que ingrese el tamaño del arreglo
tamX = int(input("Ingrese el tamaño del arreglo x: "))
tamY = int(input("Ingrese el tamaño del arreglo y: "))

# Inicializar un arreglo de ceros con el tamaño proporcionado
x = np.zeros(tamX, dtype=int)
y = np.zeros(tamY, dtype=int)


# Pedir al usuario que ingrese los elementos del arreglo
for i in range(tamX):
    elemX = int(input(f"Ingrese el elemento numero {i + 1} del arreglo x: "))
    x[i] = elemX

for j in range(tamY):
    elemY = int(input(f"Ingrese el elemento numero {j + 1} del arreglo y: "))
    y[j] = elemY
    
# Imprimir el arreglo creado
print("El arreglo x, es:", x)
print("El arreglo y, es:", y)
```

```python
# Calcula el spline cúbico natural
Spline_Cubico_Natural = SplineCubico_natural(x, y)

# Grafica los resultados
spline_cubNat_graf(Spline_Cubico_Natural, x, y)
```
