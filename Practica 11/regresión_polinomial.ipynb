```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Generamos datos aleatorios para usarlos de ejemplo.
np.random.seed(0)
X = 2*np.random.rand(200, 1)
y = 3 * X + 2 * X**2 + 5 * np.random.randn(200, 1)
```

```python
# Ajustamos una regresión polinomica de grado 2 
# NOTA: el grado del polinomio puede ser modificado segun a las necesidades del problema a tratar.
rasgos_pol = PolynomialFeatures(degree=2, include_bias=False)
X_pol = rasgos_pol.fit_transform(X)
regresion_lineal = LinearRegression()
regresion_lineal.fit(X_pol, y)
```

```python
# Haremos la vizualización de los resultados obtenidos
X_nvo = np.linspace(0, 2, 100).reshape(-1, 1)
X_nvo_pol = rasgos_pol.transform(X_nvo)
y_nvo = regresion_lineal.predict(X_nvo_pol)
```

```python
# Hacemos un grafico de los resultados anteriormente obtenidos para ver mejor como es que funciona este metodo.
plt.scatter(X, y, label='Datos de entrenamiento')
plt.plot(X_nvo, y_nvo, 'r-', label='Regresión Polinómica (grado 2)')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
